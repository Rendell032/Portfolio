## About the Dataset

### Definition of Literacy

According to UNESCO, literacy is defined as:

> The ability of a person aged 15 years and older to read, understand, and write a short, simple statement about their everyday life.

The literacy rate represents the percentage of the adult population (15+) that meets this definition.

---

## Why This Dataset Matters

Literacy is a foundational indicator of educational development and human capital. Understanding literacy trends can provide insight into broader social and economic outcomes, including education access, workforce development, and poverty reduction.

# Does literacy grow over The years?

Null Hypothesis = there is no signifigant relationship between year and global literacy rate  
Alternative hypothesis = There is a positive relationship between year nd global literacy rate
meaning global literacy increases over time

```python
# coverage analysis
entity_coverage  = (
    df.groupby('Entity')
    .agg(
        observations = ('Year', 'count'),
        first_year = ('Year','min'),
        last_year = ('Year', 'max')
    )
).reset_index()
```

```python
entity_coverage.head()
```

```python
# make calculate the life span
entity_coverage['year_span'] = entity_coverage['last_year'] - entity_coverage['first_year']

entity_coverage.sort_values(
    by =['observations', 'year_span'],
    ascending=False
).head(10)
```

```python
world = df[df['Entity'] == 'World'].copy()
```

```python
world.head()
```

we should check the coverage of world

```python
world['Year'].min(), world['Year'].max(), world['Year'].nunique()
```

```python
#check gaps
all_years = set(range(world['Year'].min(),\
                      world['Year'].max() + 1))

exsisting_years = set(world['Year'])

missing_years = sorted(all_years - exsisting_years)

len(missing_years), missing_years[:20]
```

```python
world = world.sort_values('Year')
```

```python
world.head()
```

```python
#plot the coverage
fig, ax = plt.subplots(figsize = (10,6))

ax.scatter(
    world['Year'],
    world['Literate'],
)


ax.set_title("'world' literacy Rate Over time")
ax.set_xlabel('Year')
ax.set_ylabel('literacy ratee %')
ax.grid(True)

plt.tight_layout() 
plt.show()
```

we have a lot of data that are continous from 1975 and onwards

checking correlation

```python
world[['Year','Literate']].corr()
```

```python
print('r =', world[['Year','Literate']].corr().loc['Year','Literate'])
```

looks like a very strong positive correlation

```python
x = world['Year']
y = world['Literate']

x = sm.add_constant(x)

model = sm.OLS(y, x).fit()

print(model.summary())
```

hmm should probably create a centered year variable

```python
world['year_centered'] = world['Year'] - world['Year'].mean()
world.head()
```

```python
x = sm.add_constant(world['year_centered'])
y = world['Literate']

model = sm.OLS(y,x).fit()
print(model.summary())
```

```python
lower, upper = model.conf_int().loc['year_centered']

print(f"R²: {model.rsquared:.3f}")
print(f"Slope: {model.params['year_centered']:.3f}")
print(f"P-value: {model.pvalues['year_centered']:.3e}")
print(f"95% CI: ({lower:.3f}, {upper:.3f})")
```

This regression model is to explain what there is a strong postive correlation between year and world .the model produced R² value of 94% , indicating that approximetly 94.1% of the variation in literacy rates can be explained by time. 
The slope coeffecient is 0.492 meaning that literacy rate rose increased by around 0.49 % points per year on average. The relationship is extremely significant (p < 0.001). proving a really strong evidence that literacy rates have increased over time.

```python
world['predicted'] = model.predict(x)
fig, ax = plt.subplots(figsize = (10,6))

ax.scatter(
    world['Year'],
    world['Literate'],
    label = 'Observed'
)

ax.plot(
    world['Year'],
    world['predicted'],
    label = 'Regression Line'
)

ax.set_title("'world' literacy Rate Over time")
ax.set_xlabel('Year')
ax.set_ylabel('literacy ratee %')
ax.grid(True)
ax.legend()

plt.tight_layout()
plt.show()
```

regression indecates a strong positive relationship between year and the "World" entity and literacy overtime. i assume that it represent the global literacy. however , upon visual inspection of the fitted line suggest that literacy growth may not be perfectly linear over time. 
residual pattens and the nature of the data indecate that the model should be interpreted as descriptive trend model rather than a casual or forcasting model

## Choropleth 1950 literacy percentage per country

```python
df_1950 = df_country[df_country['Year'] == 1950].copy()
geojson_url = "https://raw.githubusercontent.com/johan/world.geo.json/master/countries.geo.json"

df_1950.head()
```

```python
m = fl.Map(
    location=[20,0],
    zoom_start=1,
    tiles="OpenStreetMap",
    width=700,
    height=500
)

fl.Choropleth(
    geo_data= geojson_url,
    data = df_1950,
    columns=['Code',"Literate"],
    key_on="feature.id",
    fill_color = "RdYlGn",
    fill_opacity= 0.8,
    line_opacity=0.3,
    legend_name="Literacy Rate (1950) - percentage"
).add_to(m)

m
```
